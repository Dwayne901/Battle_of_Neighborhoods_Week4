## Analyzing the rate of vaccine distribution in the European Union

The COVID-19 pandemic in the 21st century caused a huge impact on both the economic, and the social sides of the World. Major pharmaceutical companies put forward sophisticated designs to create a vaccine against COVID-19, and so far, 6 vaccines are currently being used in the EU. The objective of this paper is to find out the current vaccine distribution rates throughout the European Union to see the number of vaccines distributed among each age group. The data is partitioned into 8 sets, the first being from the ages of 18-24, the second being 25-49, the third being 50-59, the fourth being 60-69, the fifth being 70-79, and the sixth being 80+. We also have separate partitions for healthcare workers and ‘Long Term Care Facility’(LTCF) workers. 

The data set monitors the age group, which week the vaccine was administered, type of vaccine administered, and the uptake percentages for both the first, and second doses, based on the population of the country in question. The different age ranges will show us the percentage of each age range that has gotten each dose of the vaccine, showing us the rate at which the vaccine is delivered in each country within the European Union in the time frame from the end of December 2020 to the 3rd week of February. The data in the dataset is pre-processed and normalized. We used the Frequency Distribution method, where the frequency here is the types of vaccines used in each country and how its distributed throughout the European Union. After we analyse the data, we will be able to recognize patterns describing the % uptake for each brand of vaccine in each age group, and chart graphs and models to represent them.
