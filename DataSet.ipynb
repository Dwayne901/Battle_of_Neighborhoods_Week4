# Data Section

The data set provides an overview of the progress in the rollout of COVID-19 vaccines in adults (aged 18 years and above) across the EU. It presents the number of vaccine doses distributed by manufacturers to each country for the different vaccines authorized for use in the EU and the number of first, second or unspecified doses administered to adult individuals overall, by age group and other selected target groups including healthcare workers (HCW) and residents in long-term care facilities (LTCF). 

The timeframe of our data set is from the end of December 2020 to the 3rd week of February.


The data presented in the data set is collected through The European Surveillance System (TESS). EU countries are requested to report basic indicators (number of vaccine doses distributed by manufacturers, number of first, second and unspecified doses administered) and data by target groups at national level twice a week (every Tuesday and Thursday). 

The dataset contains 5 vaccine brands, COM(Comirnaty), MOD(Moderna), AZ(AstraZeneca), SPU(Sputnik), UNK(Unknown vaccine types).


The attributes of the data set are:

1. Country

2. Age group

3. Vaccine Brand

4. Week 

5. Population of the country

6. 1st Dose and 2nd Dose of the vaccine administered


The data of the vaccine distribution is from https://www.ecdc.europa.eu/en/publications-data/covid-19-vaccine-tracker which is a free and open portal where anyone can access the data regarding the vaccine distribution in Europe. The data is available is XLS and CSV format.
